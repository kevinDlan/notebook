```python
#Definition d'une fonction
def energiePotentiel(m,h,g=9.18,EL = 150):
    E = m*h*g
    if(E>EL):
        return EL
    else:
        return E
```

```python
energiePotentiel(10,5)
```

- Exo 2

```python
def fibonacci(n):
    a = 0
    b = 1
    fib = []
    while a < n:
        fib.append(a)
        a, b = b, a+b
    return fib
```

```python
fibonacci(5)
```

```python
#Classeur
classeur = {
    "positif":[],
    "négatif":[]
}
```

```python
def trier(dic,nbre):
    if nbre > 0:
        dic['positif'].append(nbre)
    else :
        dic['négatif'].append(nbre)
    return dic
    
```

```python
trier(classeur,-10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
dataset = {f"experience{i}":np.random.randn(100,3) for i in range(6)}
```

```python
def graphe(df):
    n = len(df)
    plt.figure(figsize=(12,30))
    for k,i in zip(df.keys(), range(1, n+1)):
        plt.subplot(n,1,i)
        plt.plot(df[k])
        plt.title(k)
        
    plt.show()
```

```python
graphe(dataset)
```

```python
def scater(df):
    n = len(df)
    plt.figure(figsize=(12,30))
    for k,i in zip(df.keys(), range(1, n+1)):
        plt.subplot(n,1,i)
        plt.sca(df[k])
        plt.title(k)
        
    plt.show()
```
