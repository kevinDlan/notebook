### Visualisation avec pyplot

#### Variable qualitative

```python
# Coeur
df['coeur'].value_counts()
```

```python
df['coeur'].value_counts().plot.pie();
```

```python
df['coeur'].unique()
```

```python
plt.figure()
plt.pie(df['coeur'].value_counts(), labels=df['coeur'].unique(), autopct='%.1f%%')
plt.title('Coeur');
```

- 55,6% des individus présente une maladie cardiaque contre 44,4% qui n'en présente pas

```python
for col in df.select_dtypes(object):
    if col not in 'coeur':
        plt.figure()
        plt.pie(df[col].value_counts(), labels=df[col].unique(), autopct='%.1f%%')
        plt.title("Diagramme circulaire de " + col)
```

```python
for col in df.select_dtypes(object):
    if col not in 'coeur':
        plt.figure()
        df[col].value_counts().plot.bar()
        plt.title("Diagramme en barre de la variable " + col);
```

```python
#Age_pression
df['coeur_quant'] = df['coeur'].replace(['presence','absence'],[0,1])
df.plot.scatter('age','taux_max', c='coeur_quant', colormap='viridis')
```

- Graphe permettant de savoir si la variable age et pression peuvent nous permettre de discriminer le mieux concernant l'absence ou la présence de maladie de coeur.

- quanti vs quanti => Corrélation
- quali vs quali => test de liaison ( Indépendance ou dépendance)
- quanti vs quali => ANOVA (Analysise of Variance)
