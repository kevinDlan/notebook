### Feature engineering

- Regroupement des âges en classe d'âges

```python
df['age'].max()
```

```python
def categorisation(x):
    
    
    if x < 18:
        cate = '[0 - 18['
    elif (x >= 18) and (x < 25):
        cate = '[18 - 25['
    elif (x >= 25) and (x < 30):
        cate = '[25 - 30['
    elif (x >= 30) and (x < 35):
        cate = '[30 - 35['
    elif (x >= 35) and (x < 40):
        cate = '[35 - 40['
    elif (x >= 40) and (x < 50):
        cate = '[40 - 55['
    elif (x >= 50) and (x < 55):
        cate = '[55 - 60['
    elif (x >= 55) and (x < 60):
        cate = '[60 - 65['
    elif (x >= 60) and (x < 65):
        cate = '[65 - 70['
    elif (x >= 65) and (x < 70):
        cate = '[70 - 75['
    elif (x >= 70) and (x < 75):
        cate = '[75 - 80['
    elif (x >= 75) and (x < 80):
        cate = '[80 - 85['
    else:
        cate = '[85 - 90['
    return cate
```

```python
df['classe age'] = df['age'].map(categorisation)
```

```python
df['classe age'].value_counts().sort_values()
```

```python
plt.figure(figsize=(10, 7))
plt.bar(df['classe age'].unique(), df['classe age'].value_counts(),)
plt.xlabel('Classe d\'âge')

plt.tight_layout()
```

#### Nous sommes en présence d'une population agées dont l'âge varie entre 40 et 65 ans

### Encodage et Standardisation

```python
df1 = DF.copy()
X = df1.drop(['coeur'], axis=1)
y = df1['coeur']
```

```python
X_cat = X.select_dtypes(object)
```

```python
X_num = X.select_dtypes(exclude=object)
X_num
```

#### Encodage variable catégorielle

```python
encodage = LabelEncoder()
df1['coeur']
```

```python
def labelEnc(x):
    if x == 'presence':
        return 1
    else:
        return 0
y = df1['coeur'].map(labelEnc)
y.shape
```

```python
y.value_counts(normalize=True)*100
```

```python
for col in X_cat:
    X_cat[col] = encodage.fit_transform(X_cat[col])
X_cat
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
normaliser = StandardScaler()
X_num[X_num.columns] = normaliser.fit_transform(X_num)
X_num
```

```python
sns.boxplot(data = X_num);
```

### Concatenation

```python
X = pd.concat([X_num, X_cat], axis=1)
X
```

#### Train set et Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Régression Logistique

```python
Regression = LogisticRegression()
```

```python
Regression.fit(X_train, y_train)
```

```python
Regression.score(X_train, y_train)
```

```python
Regression.score(X_test, y_test)
```

#### Le modèle est en surapprentissage, il faut utiliser la Validation croisée.
- cross_val_score
- learning_curve(..., cv)

```python
y_pred = Regression.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
def taux_erreur(x1, x2):
    """
    Retourne
    """
    
    contingence = pd.crosstab(x1, x2)
    N = contingence.sum().sum()
    Num = contingence.iloc[1, 0] + contingence.iloc[0, 1]
    return round(Num/N * 100, 2)
```

```python
taux_erreur(y_test, y_pred)
```

#### Bagging

```python
from sklearn.svm import SVC
```

```python
bag = BaggingClassifier(base_estimator=SVC(), n_estimators=200)
```

```python
bag.fit(X_train, y_train)
```

```python
pred = bag.predict(X_test)
```

```python
confusion_matrix(y_test, pred)
```

```python
taux_erreur(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

```python
print(bag.score(X_train, y_train))
print(bag.score(X_test, y_test))
```

### Conclusion de la modélisation

- Les taux d'erreur sont les mêmes 22.22%
- Il y'a surapprentissage dans les deux cas. On peut le lever en faisant de la cross validation
